# Predicting the sale price of bulldozers

In this project I am going to predict the future sale price of bulldozers based on historical bulldozer sales data which contains several characteristics of these machines.

I will follow this framework:

![Machine learning framework](Images/6-step-ml-framework.png)

The different steps of this approach are:
   1. Problem definition
   2. Data
   3. Evaluation
   4. Features
   5. Modelling
   6. Experimentation
   
## 1. Problem definition

> How well can we predict the future sale price of a bulldozer, given its characteristics and previous examples of how much similar bulldozers have been sold for?

Since we are trying to predict a continuous value (price), it means that we are working with a **regression** problem.

## 2. Data

The data is downloaded from Kaggle's [Blue Book for Bulldozers](https://www.kaggle.com/c/bluebook-for-bulldozers/data) competition.

There are 3 datasets:

1. **Train.csv** - Historical bulldozer sales examples up to 2011 (close to 400,000 examples with 50+ different attributes, including `SalePrice` which is the **target variable**).
2. **Valid.csv** - Historical bulldozer sales examples from January 1st 2012 to April 30th 2012 (close to 12,000 examples with the same attributes as **Train.csv**).
3. **Test.csv** - Historical bulldozer sales examples from May 1st 2012 to November 2012 (close to 12,000 examples but missing the `SalePrice` attribute, as this is what we'll be trying to predict).

Since the 3 datasets are historical sales records, this shows that we are dealing with a **time series** problem. This means that there is a time attribute to the data.

So, in summary, this is a **time series** problem in which we will use **regression** to calculate the future sales price of bulldozers. 

## 3. Evaluation

For this problem, [Kaggle](https://www.kaggle.com/c/bluebook-for-bulldozers/overview/evaluation) has set the root mean squared log error (RMSLE) as the evaluation metric. The goal will be to get this value as low as possible.

To see how well my model is doing, I'll calculate the RMSLE and then compare my results to other results on the [Kaggle leaderboard](https://www.kaggle.com/c/bluebook-for-bulldozers/leaderboard).

## 5. Modelling

Since this is a regression problem, I will be using the **Random Forest** algorithm.

Before I start doing anything, I have to split the data that I have been working with into training and validation sets. This is because I have been working on data that comes from the combined training and validation sets. 

I can split the data by the dates that are referred in the beginning of this notebook. The validation set contains all the sales data from 2012 onwards while the training set contains all the data from before 2012.

```python
# Splitting data into training and validation sets
data_train = df_temp[df_temp['saleYear'] != 2012]
data_cv = df_temp[df_temp['saleYear'] == 2012]

# Check number of examples in each set
len(data_train), len(data_cv)
```

```python
# Split data into X and y in training and validation sets
X_train = data_train.drop('SalePrice', axis=1)
y_train = data_train['SalePrice']
X_cv = data_cv.drop('SalePrice', axis=1)
y_cv = data_cv['SalePrice']
```

I will build a function to fit the model to the data, make predictions and evaluate the model with RMSLE.

```python
def evaluate_model(model, X_train, X_cv, y_train, y_cv):
    '''
    Fits the model and makes predictions on the training and validation set.
    Then calculates root mean squared log error between predictions and true labels.
    '''
    # Fit model to the training set
    model.fit(X_train, y_train)
    
    # Predictions on training and validation sets
    train_preds = model.predict(X_train)
    cv_preds = model.predict(X_cv)
    
    # Calculate rmsle
    # To get rmsle, just need to compute the square root of the mean squared log error
    rmsle_train = np.sqrt(mean_squared_log_error(y_train, train_preds))
    rmsle_cv = np.sqrt(mean_squared_log_error(y_cv, cv_preds))
    
    score = {'Training RMSLE': rmsle_train,
             'Validation RMSLE': rmsle_cv}
    
    return score
```

```python
%%time

# Instantiate Random Forest model
rf = RandomForestRegressor(n_jobs=-1, random_state=42)

# Evaluate Random Forest model
evaluate_model(rf, X_train, X_cv, y_train, y_cv)
```

The RMSLE score in the validation set is already pretty good but it took 6m44s to run the evaluation function.

Given that the training set has 400.000 examples and training the model on a dataset of this size takes quite a while, I will use a subset of 10.000 examples to train the model.

This way I can keep experimenting without having to wait very long every time I retrain the model.

```python
%%time

# Instantiate Random Forest model
rf = RandomForestRegressor(n_jobs=-1, random_state=42, max_samples=10000)

# Evaluate Random Forest model
evaluate_model(rf, X_train, X_cv, y_train, y_cv)
```

### Hyperparameter tuning with RandomizedSearchCV

I will tune the hyperparameters on the subset and then train the model on the full training set with the best hyperparameters found.

Afterwards, I will tune the hyperparameters on the full training set, train the model and compare both rmsle scores.

**Subset**

```python
%%time

# Hyperparameter grid for Random Forest
rf_sub_grid = {'n_estimators': np.arange(10, 100, 10),
           'max_depth': [None, 3, 5, 10],
           'min_samples_split': np.arange(2, 20, 2),
           'min_samples_leaf': np.arange(1, 20, 2),
           'max_features': [0.5, 1, 'sqrt', 'auto'],
           'max_samples': [10000]}

# Instantiate RandomizedSearchCV model
rs_sub_rf = RandomizedSearchCV(RandomForestRegressor(),
                           param_distributions=rf_sub_grid,
                           n_iter=10,
                           cv=5,
                           verbose=True)

# Fit model to the data
rs_sub_rf.fit(X_train, y_train)
```

```python
# Best hyperparameters found
best_params = rs_sub_rf.best_params_
best_params
```

```python
# Model with best hyperparameters found
model = RandomForestRegressor(n_estimators=60, 
                              min_samples_split=4,
                              min_samples_leaf=11,
                              max_features='auto',
                              max_depth=None,
                              n_jobs=-1)

# Evaluate model
evaluate_model(model, X_train, X_cv, y_train, y_cv)
```

**Full training set**

```python
%%time

# Hyperparameter grid for Random Forest
rf_grid = {'n_estimators': np.arange(10, 100, 10),
           'max_depth': [None, 3, 5, 10],
           'min_samples_split': np.arange(2, 20, 2),
           'min_samples_leaf': np.arange(1, 20, 2),
           'max_features': [0.5, 1, 'sqrt', 'auto']}

# Instantiate RandomizedSearchCV model
rs_rf = RandomizedSearchCV(RandomForestRegressor(),
                           param_distributions=rf_grid,
                           n_iter=100,
                           cv=5,
                           verbose=True)

# Fit model to the data
rs_rf.fit(X_train, y_train)
```

```python
# Best hyperparameters found
best_params = rs_rf.best_params_
best_params
```

```python
# Model with best hyperparameters found
model = RandomForestRegressor(n_estimators=70, 
                              min_samples_split=16,
                              min_samples_leaf=1,
                              max_features=0.5,
                              max_depth=None,
                              n_jobs=-1)

# Evaluate model
evaluate_model(model, X_train, X_cv, y_train, y_cv)
```

The hyperparameter tuning on the subset took almost 7 minutes while the hyperparameter tuning on the whole training set took close to 7 hours. When tuning the hyperparameters on the subset, the model improved its rmsle score on the validation set. As expected, the hyperparameter tuning on the whole training set got better results and the model performed better than with the hyperparameter tuning on the subset.

Experimenting with a subset is much quicker and I can run RandomizedSearchCV several times and compare different hyperparameters found to see which ones improve the model and which ones don't. Sometimes we can get lucky and find a great set of hyperparameters.

Below are the best set of hyperparameters that I found and funnily enough, I found them when running RandomizedSearchCV on the subset. I ran RandomizedSearchCV on the subset several times and compared the hyperparameters found.

```python
# Model with best hyperparameters found
model = RandomForestRegressor(n_estimators=40, 
                              min_samples_split=8,
                              min_samples_leaf=3,
                              max_features=0.5,
                              max_depth=None,
                              n_jobs=-1)

# Evaluate model
evaluate_model(model, X_train, X_cv, y_train, y_cv)
```

### Make Predictions on test set

Before using the test set to make predictions, I will have to use the same preprocessing steps that I used for the training and validation sets.

To do this, I will create a function that runs all these steps.

```python
def preprocessing(df):
    '''
    Performs the same preprocessing steps as the ones done for the training and validation sets
    '''
    # Add new features to dataset
    df['saleYear'] = df['saledate'].dt.year
    df['saleMonth'] = df['saledate'].dt.month
    df['saleDay'] = df['saledate'].dt.day
    df['saleDayOfWeek'] = df['saledate'].dt.dayofweek
    df['saleDayOfYear'] = df['saledate'].dt.dayofyear
    df['saleEndOfMonth'] = df['saledate'].dt.is_month_end
    df['saleStartOfMonth'] = df['saledate'].dt.is_month_start
    df['saleStartOfQuarter'] = df['saledate'].dt.is_quarter_start
    df['saleEndOfQuarter'] = df['saledate'].dt.is_quarter_end
    df['saleStartOfYear'] = df['saledate'].dt.is_year_end
    df['saleEndOfYear'] = df['saledate'].dt.is_year_start

    # Since I created all these features, I don't need the 'saledate' variable anymore
    df.drop('saledate', axis=1, inplace=True)
    
    for column_label, values in df.items():
        # Find the columns which contain 'string' datatype and turn them into 'category' datatypes
        if pd.api.types.is_string_dtype(values):
            df[column_label] = values.astype('category').cat.as_ordered()
            # Turn categories into numbers (missing value will get value 0)
            # We add the +1 because Pandas encodes missing categories as -1
            df[column_label] = pd.Categorical(values).codes + 1
            
        # Find columns with 'numeric' datatype and missing values and fill the missing values with the mode of each column
        elif pd.api.types.is_numeric_dtype(values):
            if pd.isna(values).sum():
                # Fill missing values with median
                df[column_label] = values.fillna(values.mode()[0]) 
```

```python
# Import test set
df_test = pd.read_csv('Data/Test.csv', low_memory=False, parse_dates=['saledate'])

# Check dimensions of test set
df_test.shape
```

```python
# Preprocess test set
preprocessing(df_test)

# Check dimensions of test set
df_test.shape
```

```python
# View test set
df_test.head().T
```

```python
# Instantiate model with best hyperparameters found
rf_model = RandomForestRegressor(n_estimators=40, 
                              min_samples_split=8,
                              min_samples_leaf=3,
                              max_features=0.5,
                              max_depth=None,
                              n_jobs=-1)

# Fit model to the data
rf_model.fit(X_train, y_train)

# Make predictions
y_preds = rf_model.predict(df_test)
```
